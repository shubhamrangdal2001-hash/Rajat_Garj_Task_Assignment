```python
# Mini BPE implementation
words = {
"l o w": 5,
"l o w e r":2
}
def get_status(words):
  lst = []
  for i , n in words.items():
    for j in i:
        if j!=" ":
          for k in range(n):
            lst.append(j)
  dct = {}
  lst1 = list(set(lst))
  for i in range(len(lst1)):
    for j in range(i+1,len(lst)-i):
      dct[(lst1[i], lst[j])] = dct.get((lst1[i], lst[j]),0) + 1
  return dct
get_status(words)
```

```python
def get_stats(words):
    pairs = {}

    for word, freq in words.items():
        s = word.split()
        for i in range(len(s) - 1):
            pair = (s[i], s[i+1])
            if pair not in pairs:
                pairs[pair] = 0
            pairs[pair] += freq
    return pairs
words = {
    "l o w": 5,
    "l o w e r": 2,
}

print(get_stats(words))
```
