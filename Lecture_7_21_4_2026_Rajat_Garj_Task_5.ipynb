```python
text = [
    "book a  flight",
    "cancel my ticket",
    "weather today",
    "book hotel",
    "cancel reservation",
    "waether tomorrow"
]
labels = ["booking" , "cancel" , "weather", "booking","cancel","weather"]
import pandas as pd

def tokenizer_simple(text):
  return text.split()
def tokenizer_custom(text_item):
  import re
  text_item = text_item.lower()
  text_item = re.sub("r[^a-z ]" , "" , text_item)
  return text_item.split()
token = [tokenizer_custom(item) for item in text]
print(token)
```

```python
from numpy import vectorize
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
X_test =  [
    "bok a  flight",
    "cancel my ticket",
    "weather today",
    "book hotel",
    "cancel reservation",
    "waether tomorrow"
]
def run_pipeline(tokenizer):
  vec = TfidfVectorizer(tokenizer= tokenizer)
  X = vec.fit_transform(text)
  model = LogisticRegression()
  model.fit(X, labels)
  preds = model.predict(vec.transform(X_test)) # Transformed X_test using the fitted vectorizer
  print(classification_report(labels , preds)) # Note: 'labels' here are from training data, might need 'labels_test' for proper test set evaluation
  print(preds)
run_pipeline(tokenizer_custom)
run_pipeline(tokenizer_custom)
```
